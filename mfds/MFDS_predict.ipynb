```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data_set.csv")
df
```

```python
teams = df["team 1"].unique()
teams
```

```python
dit = {}
for i in teams:
    a = df[df["team 1"] == i]
    dit[i] = [a["runs 1"].mean(),a["runs 1"].sum()/a["runs 1"].mean()]
    
```

```python
dit #averages of teams batting 1st, no. of matches played
```

```python
dit2 = {}
for i in teams:
    a = df[df["team 2"] == i]
    dit2[i] = [a["runs2_projected"].mean(),a["runs2_projected"].sum()/a["runs2_projected"].mean()]
    
```

```python
dit2  #averages of teams batting 2nd, no. of matches played
```

```python
dit.keys()
```

```python
dit['MI'][0]
```

```python
teams_avg = {}
for i in teams:
    teams_avg[i] = ((dit[i][0]*dit[i][1])+(dit2[i][0]*dit2[i][1]))/(dit[i][1]+dit2[i][1])
```

```python
teams_avg
```

```python
pitch_avg = {}
pitches = df['stadium'].unique()
pitches
```

```python
for i in pitches:
    df_pitch = df[df["stadium"] == i]
    pitch_avg[i]  = df_pitch['runs 1'].mean()
   
```

```python
pitch_avg
```

```python
bowling_average = {}
for i in teams:
    df_bowlings = df[df["team 2"] == i]
    bowling_average[i] = [df_bowlings['runs 1'].mean(),df_bowlings['runs 1'].sum()/df_bowlings['runs 1'].mean()]
```

```python
bowling_average
```

```python
bowling_average_2 = {}
for i in teams:
    df_bowlings = df[df["team 1"] == i]
    bowling_average_2[i] = [df_bowlings['runs 2'].mean(),df_bowlings['runs 2'].sum()/df_bowlings['runs 2'].mean()]
```

```python
bowling_average_2
```

```python
bowling_avg = {}
for i in teams:
    bowling_avg[i] = ((bowling_average[i][0]*bowling_average[i][1])+(bowling_average_2[i][0]*bowling_average_2[i][1]))/(bowling_average[i][1]+bowling_average_2[i][1])
```

```python
bowling_avg
```
